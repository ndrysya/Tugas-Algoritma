```python
# Latihan 5.21.6 Menyaring data menggunakan pandas

# Menampilkan data dengan nilai 'petal length (cm)' == 3.5
data_filtered = data_csv[data_csv['petal length (cm)'] == 3.5]
print (data_filtered)

# Menampilkan data yang memiliki nilai 'sepal width (cm)' lebih besar dari 3
data_filtered2 = data_csv[data_csv['sepal width (cm)'] > 3]
print (data_filtered2)

# Menampilkan data yang memenuhi dua kondisi
data_filtered3 = data_csv[(data_csv['petal length (cm)'] == 3.5) & (data_csv['sepal width (cm)'] > 3)]
print (data_filtered3)

# Menampilkan data yang memenuhi salah satu dari dua kondisi
data_filtered4 = data_csv[(data_csv['petal length (cm)'] == 3.5) | (data_csv['sepal width (cm)'] > 3)]
print (data_filtered4)

# Menampilkan data dengan menggunakan method loc
data_filtered5 = data_csv.loc[(data_csv['petal length (cm)'] == 3.5) & (data_csv['sepal width (cm)'] > 3), ['petal length (cm)', 'sepal width (cm)']]
print (data_filtered5)

# Menampilkan data yang memiliki nilai NaN pada kolom 'petal length (cm)'
data_filtered6 = data_csv[pd.isnull(data_csv['petal length (cm)'])]
print (data_filtered6)
```

```python
# Latihan 5.21.7 Pengelompokan data menggunakan pandas


# Mengelompokkan data berdasarkan kolom 'sepal width (cm)'
grouped_data = data_csv.groupby('sepal width (cm)')
# Menghitung rata-rata pada kolom 'petal length (cm)' untuk setiap kelompok
avg_petal_length = grouped_data['petal length (cm)'].mean()
print(avg_petal_length)

#Pandas menyediakan fungsi groupby() untuk melakukan pengelompokan data berdasarkan satu atau beberapa kolom pada DataFrame
#Outputnya akan menampilkan rata-rata panjang petal length (cm) untuk setiap kelompok sepal width  pada DataFrame data_csv
```

```python
# Mengelompokkan data berdasarkan kolom 'sepal width (cm)' dan 'petal width (cm)'
grouped_data = data_csv.groupby(['sepal width (cm)', 'petal width (cm)'])

# Menghitung jumlah data pada setiap kelompok
count_data = grouped_data.size()

# Menampilkan hasil pengelompokan
print(count_data)

#Outputnya akan menampilkan jumlah data untuk setiap kelompok yang terbentuk dari kombinasi nilai pada 
#kolom sepal width (cm) dan petal width (cm) pada DataFrame data_csv
```

```python
# Latihan 5.22.4 Properti pembuatan histogram dengan matplotlib
import matplotlib.pyplot as plt
import pandas as pd

fig, ax = plt.subplots(figsize=(8,6))
ax.hist(data_csv['petal length (cm)'], bins=20, alpha=0.8, edgecolor='black', hatch='//')

ax.set_xlabel('Petal Length (cm)',fontsize=14, fontweight='bold')
ax.set_ylabel('Frekuensi', fontsize=14, fontweight='bold')
ax.set_title('Distribusi Panjang Mahkota Bunga Iris', fontname = 'Times New Roman', fontsize=16, fontweight='bold')

ax.tick_params(axis='both', which='major', labelsize=12)
ax.grid(axis='y', alpha=0.5)

plt.show()

# Perbedaan dengan contoh sebelumnya adalah penambahan parameter hatch pada fungsi ax.hist(), 
# yang diisi dengan string pola/hatch yang ingin digunakan pada bar histogram. 
# Dalam contoh di atas, pola yang digunakan adalah //. Selain itu, contoh di atas juga
# menambahkan parameter alpha untuk menentukan tingkat transparansi bar, dan parameter edgecolor untuk memberikan warna pada garis tepi ba
```
